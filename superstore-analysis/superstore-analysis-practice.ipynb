```python
import pandas as pd
df = pd.read_csv("raw_superstore_data.csv")
df.head()
```

```python
#Check number of rows and columns
print ("Numbersof rows:", df.shape[0])
print ("Number of columns:", df.shape[1])

#Display column names
print ("Column Names:", df.columns)
```

```python
#Check for missing values

missing_transactionid = df["Transaction_ID"].isnull().sum()
missing_country = df["Country"].isnull().sum()
missing_region = df["Region"].isnull().sum()
missing_storetype = df["Store_Type"].isnull().sum()
missing_category = df["Category"].isnull().sum()
missing_itemid = df["Item_ID"].isnull().sum()
missing_quantity = df["Quantity"].isnull().sum()
missing_priceperunit = df["Price_per_Unit"].isnull().sum()
missing_totalsales = df["Total_Sales"].isnull().sum()
missing_discountapplied = df["Discount_Applied"].isnull().sum()
missing_day = df["Day"].isnull().sum()
missing_timeofpurchase = df["Time_of_Purchase"].isnull().sum()
missing_paymentmethod = df["Payment_Method"].isnull().sum()
missing_customerid = df["Customer_ID"].isnull().sum()
missing_loyaltycard = df["Loyalty_Card_Used"].isnull().sum()
missing_employeeid = df["Employee_ID"].isnull().sum()
missing_feedbackscore = df["Feedback_Score"].isnull().sum()
missing_return = df["Return"].isnull().sum()
missing_weather = df["Weather"].isnull().sum()
missing_holiday = df["Holiday"].isnull().sum()

#Print missing values for each column
print ("Missing values in Transaction ID:", missing_transactionid)
print ("Missing values in Country:", missing_country)
print ("Missing values in Region:", missing_region)
print ("Missing values in Store_Type:", missing_storetype)
print ("Missing values in Category:", missing_category)
print ("Missing values in Item_ID:", missing_itemid)
print ("Missing values in Quantity:", missing_quantity)
print ("Missing values in Price_per_Unit:", missing_priceperunit)
print ("Missing values in Total_Sales:", missing_totalsales)
print ("Missing values in Discount_Applied:", missing_discountapplied)
print ("Missing values in Day:", missing_day)
print ("Missing values in Time_of_Purchase:", missing_timeofpurchase)
print ("Missing values in Payment_Method:", missing_paymentmethod)
print ("Missing values in Customer_ID:", missing_customerid)
print ("Missing values in Loyalty_Card_Used:", missing_loyaltycard)
print ("Missing values in Employee_ID:", missing_employeeid)
print ("Missing values in Feedback_Score:", missing_feedbackscore)
print ("Missing values in Return:", missing_return)
print ("Missing values in Weather:", missing_weather)
print ("Missing values in Holiday:", missing_holiday)
```

```python
#Removing whitespaces from specifc columns

df["Country"] = df["Country"].str.strip()
df["Region"] = df["Region"].str.strip()
df["Store_Type"] = df["Store_Type"].str.strip()
df["Category"] = df["Category"].str.strip()
df["Day"] = df["Day"].str.strip()
df["Payment_Method"] = df["Payment_Method"].str.strip()
df["Weather"] = df["Weather"].str.strip()

print ("Whitespace removed from text columns.")

df.head()
```

```python
df.shape
```

```python
#total rows before removing duplicates
rows_before = df.shape[0]
dupes = df.duplicated().sum()
#remove duplicates
df = df.drop_duplicates()

#total rows after removing duplicates
rows_after = df.shape[0]

print ("Rows before removing duplicates:", rows_before)
print ("Rows after removing duplicates:", rows_after)
print ("Numner od Duplicate Rows:", dupes)
```

```python
#Define a region mapping

region_mapping = {
    'USA': 'North',
    'India': 'Central',
    'Brazil': 'South',
    'China': 'East',
    'Germany': 'West', 
    'Australia': 'East',
    'South Africa': 'South',
    'Canada': 'North',
    'UK': 'West',
    'Japan': 'East'
}

df['Region'] = df.apply(
    lambda row: region_mapping[row['Country']] if pd.isnull(row['Region']) else row['Region'],
    axis=1
)  

df.head(30)
```

```python
avg_price_per_category = df.groupby('Category')['Price_per_Unit'].mean()

df.loc[(df['Category'] == 'Fruits') & 
       (df['Price_per_Unit'].isnull()), 
       'Price_per_Unit'] = avg_price_per_category['Fruits']

df.loc[(df['Category'] == 'Vegetables') & 
       (df['Price_per_Unit'].isnull()), 
       'Price_per_Unit'] = avg_price_per_category['Vegetables']

df.loc[(df['Category'] == 'Dairy') & 
       (df['Price_per_Unit'].isnull()), 
       'Price_per_Unit'] = avg_price_per_category['Dairy']

df.loc[(df['Category'] == 'Beverages') & 
       (df['Price_per_Unit'].isnull()), 
       'Price_per_Unit'] = avg_price_per_category['Beverages']

df.loc[(df['Category'] == 'Meat') & 
       (df['Price_per_Unit'].isnull()), 
       'Price_per_Unit'] = avg_price_per_category['Meat']

df.loc[(df['Category'] == 'Electronics') & 
       (df['Price_per_Unit'].isnull()), 
       'Price_per_Unit'] = avg_price_per_category['Electronics']

df.loc[(df['Category'] == 'Household') & 
       (df['Price_per_Unit'].isnull()), 
       'Price_per_Unit'] = avg_price_per_category['Household']

df.loc[(df['Category'] == 'Snacks') & 
       (df['Price_per_Unit'].isnull()), 
       'Price_per_Unit'] = avg_price_per_category['Snacks']

df.loc[(df['Category'] == 'Clothing') & 
       (df['Price_per_Unit'].isnull()), 
       'Price_per_Unit'] = avg_price_per_category['Clothing']

df.loc[(df['Category'] == 'None') & 
       (df['Price_per_Unit'].isnull()), 
       'Price_per_Unit'] = avg_price_per_category['None']


df.isnull().sum()

```

```python
df['Total_Sales'] = df['Quantity'] * df['Price_per_Unit']
```

```python
common_store_type_per_region = df.groupby(
    'Region')['Store_Type'].agg(lambda x: x.mode()[0])

df.loc[(df['Region'] == 'North') & 
       (df['Store_Type'].isnull()), 
       'Store_Type'] = common_store_type_per_region['North']

df.loc[(df['Region'] == 'Central') & 
       (df['Store_Type'].isnull()), 
       'Store_Type'] = common_store_type_per_region['Central']

df.loc[(df['Region'] == 'South') & 
       (df['Store_Type'].isnull()), 
       'Store_Type'] = common_store_type_per_region['South']

df.loc[(df['Region'] == 'West') & 
       (df['Store_Type'].isnull()), 
       'Store_Type'] = common_store_type_per_region['West']

df.loc[(df['Region'] == 'East') & 
       (df['Store_Type'].isnull()), 
       'Store_Type'] = common_store_type_per_region['East']

df.isnull().sum()
```

```python
avg_feedback_per_store = df.groupby(
    'Store_Type')['Feedback_Score'].mean()

 
df.loc[(df['Store_Type'] == 'Supermarket') &
       (df['Feedback_Score'].isnull()), 
       'Feedback_Score'] = avg_feedback_per_store['Supermarket']

df.loc[(df['Store_Type'] == 'Grocery Store') &
       (df['Feedback_Score'].isnull()), 
       'Feedback_Score'] = avg_feedback_per_store['Grocery Store']

df.loc[(df['Store_Type'] == 'Hypermarket') &
       (df['Feedback_Score'].isnull()), 
       'Feedback_Score'] = avg_feedback_per_store['Hypermarket']

df.loc[(df['Store_Type'] == 'Wholesale') &
       (df['Feedback_Score'].isnull()), 
       'Feedback_Score'] = avg_feedback_per_store['Wholesale']

df.loc[(df['Store_Type'] == 'None') &
       (df['Feedback_Score'].isnull()), 
       'Feedback_Score'] = avg_feedback_per_store['None']

df.isnull().sum()
```

```python
common_payment_per_store = df.groupby(
    'Store_Type')['Payment_Method'].agg(lambda x: x.mode()[0])

 
df.loc[(df['Store_Type'] == 'Supermarket') & 
       (df['Payment_Method'].isnull()), 
       'Payment_Method'] = common_payment_per_store['Supermarket']

df.loc[(df['Store_Type'] == 'Grocery Store') &
       (df['Payment_Method'].isnull()), 
       'Payment_Method'] = common_payment_per_store['Grocery Store']

df.loc[(df['Store_Type'] == 'Wholesale') &
       (df['Payment_Method'].isnull()), 
       'Payment_Method'] = common_payment_per_store['Wholesale']

df.loc[(df['Store_Type'] == 'Hypermarket') &
       (df['Payment_Method'].isnull()), 
       'Payment_Method'] = common_payment_per_store['Hypermarket']

df.isnull().sum()
```
