```python
# pop() Removes and returns the item at a specified index (or the last item by default).
last_item = my_list.pop()
print(last_item)  # Output: 8
print(my_list)    # Output: [1, 2, 3, 4, 5, 6, 7]
```

```python
# clear () remove all items from the list 
my_list.clear()
print(my_list)  # Output: []
```

```python
# index () Returns the index of the first item with a specified value.
my_list = [1, 2, 3, 4, 5]
index_of_3 = my_list.index(3)
print(index_of_3)  # Output: 2
```

```python
# count() Returns the number of times a specified value appears in the list.
my_list = [1, 2, 3, 2, 4, 2]
count_of_2 = my_list.count(2)
print(count_of_2)  # Output: 3
```

```python
# sort() Sorts the list in ascending order. It can also sort in descending order if specified.
my_list = [3, 1, 4, 5, 2]
my_list.sort()
print(my_list)  # Output: [1, 2, 3, 4, 5]

my_list.sort(reverse=True)
print(my_list)  # Output: [5, 4, 3, 2, 1]
```

```python
# copy() Returns a shallow copy of the list.
my_list=[1,2,3,4,5]
new_list = my_list.copy()
print(new_list)  # Output: [1, 2, 3, 4, 5]
```

```python
# lenght of list
print(len(my_list))  # Output: 5
```

```python
# min max in the list
print(max(my_list))  # Output: 5
print(min(my_list))  # Output: 1
```

```python
# sum(): Returns the sum of all elements in the list (works only for lists of numbers).
print(sum(my_list))  # Output: 15
```

```python
# combining list method You can also chain or combine list methods to perform multiple operations at once. For example:
my_list = [5, 3, 2, 8, 1]
my_list.sort()
my_list.append(10)
print(my_list)  # Output: [1, 2, 3, 5, 8, 10]
```

# python if else

```python
a = 33
b = 200
if b > a:
  print("b is greater than a")
```

```python
# The elif keyword is Python's way of saying "if the previous conditions were not true, then try this condition".
a = 33
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
```

```python
# The else keyword catches anything which isn't caught by the preceding conditions.
a = 200
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

```python
# You can also have an else without the elif:
a = 200
b = 33
if b > a:
  print("b is greater than a")
else:
  print("b is not greater than a")
```

```python
# Short Hand If
# If you have only one statement to execute, you can put it on the same line as the if statement.
if a > b: print("a is greater than b")
```

```python
# Short Hand If ... Else
# If you have only one statement to execute, one for if, and one for else, you can put it all on the same line:
a = 2
b = 330
print("A") if a > b else print("B")
```

```python
# One line if else statement, with 3 conditions:
a = 330
b = 330
print("A") if a > b else print("=") if a == b else print("B")
```

```python
# The and keyword is a logical operator, and is used to combine conditional statements:
# Test if a is greater than b, AND if c is greater than a:
a = 200
b = 33
c = 500
if a > b and c > a:
  print("Both conditions are True")
```

```python
# The or keyword is a logical operator, and is used to combine conditional statements:
# Test if a is greater than b, OR if a is greater than c:
a = 200
b = 33
c = 500
if a > b or a > c:
  print("At least one of the conditions is True")
```

```python
# Example 1: Checking Age and Citizenship
# Suppose we want to check if someone is eligible to vote based on their age and citizenship.
age = 20
citizen = True

if age >= 18:
    if citizen:
        print("You are eligible to vote!")
    else:
        print("You must be a citizen to vote.")
else:
    print("You must be at least 18 years old to vote.")
# Explanation:
# The outer if checks if the person is at least 18 years old.
# If true, the inner if checks if the person is a citizen.
# If both conditions are true, it prints that the person is eligible to vote; otherwise, it gives a specific reason.
```

```python
# Example 2: Checking Grade Levels
# check if a student’s grade falls into different ranges.
score = 85

if score >= 70:
    if score >= 90:
        print("Grade: A")
    elif score >= 80:
        print("Grade: B")
    else:
        print("Grade: C")
else:
    print("Grade: F")
# Explanation:
# The outer if checks if the score is at least 70.
# Inside this if, we use nested conditions to determine if the score qualifies for A, B, or C grades.
# If the score is below 70, it directly assigns a grade of F.
```

```python
# Example 3: Account Login with Nested Checks
# This example shows nested if statements for a basic login check, verifying both username and password.
username = "admin"
password = "12345"

if username == "admin":
    if password == "12345":
        print("Login successful!")
    else:
        print("Incorrect password.")
else:
    print("Username not found.")

#Explanation:
#The outer if checks if the username is correct.
#If the username is correct, the inner if checks if the password is correct.
#If both conditions are met, it prints "Login successful!"
```

# 2 while loop

```python
# Simple Example: Counting Down with a while Loop
count = 5
while count > 0:
    print("Counting down:", count)
    count -= 1  # Decreases count by 1 each time
```

```python
# Example 3: Simple Addition Game
# This loop will keep asking the user to add two numbers until they get the answer right.
answer = 0
while answer != 7: 
    answer = int(input("What is 3 + 4? "))
print("Correct! Well done.")
# answer != 7: This is the loop’s condition. It means “keep looping as long as answer is not equal to 7.”
```

```python
# Example 2: Asking for Password
# This loop will keep asking the user for the correct password until they enter it correctly.
password = ""
while password != "python123":
    password = input("Enter the password: ")
print("Access granted!")
```

```python
# Example 3: Guess the Number Game
# This example demonstrates a simple guessing game where the user tries to guess a number between 1 and 10.
import random
secret_number = random.randint(1, 10)
guess = None

while guess != secret_number:
    guess = int(input("Guess the number between 1 and 10: "))
    if guess < secret_number:
        print("Too low, try again!")
    elif guess > secret_number:
        print("Too high, try again!")
    else:
        print("Congratulations! You guessed it!")
```

```python
# Example: Stopping Early with break
for i in range(10):
    if i == 5:
        break
    print("Number:", i)
```

```python
# Example: Skipping Even Numbers with continue
for i in range(6):
    if i % 2 == 0:  # Checks if `i` is an even number % means reminder
        continue    # Skips the rest of the loop for even numbers
    print("Odd number:", i)
```

```python
# Example 4: Function with Default Parameters
# Sometimes we want a parameter to have a default value in case it isn’t provided.
def greet_person(name="Guest"):
    print("Hello, " + name + "!")
    
greet_person("Alice")  # Uses "Alice" as the name
greet_person()         # Uses the default value "Guest"
# Explanation: If we call greet_person() without a parameter, it defaults to "Guest". Otherwise, it uses the provided name.
```

```python
# Example 5: Returning a Value
# A function can return a value to the code that called it using the return statement.
def square(number):
    return number * number

result = square(4)
print("The square of 4 is:", result)
# Explanation: The square function takes a number and returns its square. We capture the result in result and print it.
```

```python
# Example 6: Using Functions Inside Other Functions
# Functions can call other functions, allowing us to break down complex tasks.
def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

def calculate(a, b):
    sum_result = add(a, b)
    product_result = multiply(a, b)
    print("Sum:", sum_result)
    print("Product:", product_result)
    
calculate(3, 5)
```

# How numpy is helpful in data analysis

# 1. Handling Large Datasets Efficiently

```python
# Example 
# generate 1000000 random number and calculate mean and std-deviation
import numpy as np

# Creating a large array of random data
data = np.random.rand(1000000)

# Calculating the mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Standard Deviation:", std_dev)
```

# 2. Cleaning Data

```python
# Data cleaning often involves removing or replacing missing values. NumPy’s functions make this easy.
import numpy as np

# Simulating a dataset with missing values (NaNs)
data = np.array([1, 2, np.nan, 4, np.nan, 6])

# Finding and counting missing values
missing_values = np.isnan(data)
count_missing = np.sum(missing_values)
print("Missing values:", count_missing)

# Replacing missing values with the mean
mean_value = np.nanmean(data)  # Ignores NaNs while calculating mean
data_cleaned = np.where(np.isnan(data), mean_value, data)
print("Data after cleaning:", data_cleaned)
```

# 3. Filtering Data

```python
# Filtering helps in selecting relevant data points. NumPy’s logical 
# operations make filtering data intuitive.
import numpy as np

# Generating a sample dataset
data = np.array([5, 12, 7, 10, 15, 3, 8])

# Filter to find values greater than 10
filtered_data = data[data > 10]
print("Values greater than 10:", filtered_data)

# Filter to find values between 5 and 10
filtered_range = data[(data >= 5) & (data <= 10)]
print("Values between 5 and 10:", filtered_range)
```

# 4. Statistical Analysis

```python
import numpy as np

# Simulating two related datasets
data1 = np.random.normal(50, 5, 1000)  # Dataset 1 with mean=50 and std=5
data2 = np.random.normal(0, 2, 1000)  # Dataset 2 with mean=0 and std=2

# Calculating correlation between two datasets
correlation = np.corrcoef(data1, data2)
print("Correlation coefficient:\n", correlation)
```

```python
# Calculating quantiles
quantiles = np.percentile(data1, [25, 50, 75])
print("25th, 50th, and 75th percentiles:", quantiles)
```

# 5. Data Aggregation and Grouping

```python
# NumPy’s aggregation functions help summarize data, like calculating group-level statistics.
import numpy as np

# Simulating grouped data (e.g., daily temperatures over 7 days)
data = np.array([[72, 75, 78], [68, 70, 72], [60, 62, 65]])

# Calculating the mean temperature for each day (row-wise)
mean_temp_per_day = np.mean(data, axis=1)
print("Mean temperature per day:", mean_temp_per_day)

# Calculating the mean temperature per hour (column-wise)
mean_temp_per_hour = np.mean(data, axis=0)
print("Mean temperature per hour:", mean_temp_per_hour)
```

# 6. Working with Large Datasets for Machine Learning

```python
# NumPy is heavily used to prepare datasets for machine learning models, including feature scaling and normalizing.
import numpy as np

# Simulating a dataset of features with varying scales
data = np.array([[1, 100], [2, 200], [3, 300]])

# Feature scaling (min-max normalization)
data_min = np.min(data, axis=0)
data_max = np.max(data, axis=0)
scaled_data = (data - data_min) / (data_max - data_min)
print("Scaled data:\n", scaled_data)
```

```python
# Normalizing (making mean=0 and std=1)
mean = np.mean(data, axis=0)
std = np.std(data, axis=0)
normalized_data = (data - mean) / std
print("Normalized data:\n", normalized_data)
```

# 6. Subplots

```python
# Subplots allow multiple plots to be displayed within a single figure, useful for comparing data.
# Creating data for plots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Plotting in first subplot
ax1.plot(x, y1, color="blue")
ax1.set_title("Sine Wave")

# Plotting in second subplot
ax2.plot(x, y2, color="red")
ax2.set_title("Cosine Wave")

plt.suptitle("Sine and Cosine Waves")
plt.show()
```

# 7. Customizing Plots
