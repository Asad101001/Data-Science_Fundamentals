```python
#Seaborn:

#SCATTERPLOT

import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("Tips")

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
plt.title("Scatterplot of Total Bill vs Tip")
plt.show()
tips.head()
```

```python
#LINEPLOT

fmri  = sns.load_dataset("fmri")

sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event")
plt.title("FMRI Signal Over Time")
plt.show()
```

```python
#HISTOGRAM and KERNEL DENSIY ESTIMATE(K.D.E) plot

sns.histplot(tips["total_bill"], kde=True , color="indigo")
plt.title("Histogram and KDE of Total Bill")
plt.show()
```

```python
#BAR PLOT
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("Tips")

sns.barplot(data=tips, x="day", y="tip", estimator=sum,
            ci=None, palette="viridis")
plt.title("Total Tips By Day")
plt.show()
```

```python
#BOX plot
# display the distribution of data and highlight
# can show the following values : min,max,median,q1,q3,outliers

sns.boxplot(data=tips, x="day", y="total_bill", palette="coolwarm")
plt.title("Total Bill Distribution by Day")
plt.show()
```

```python
#VIOLIN PLOT
# shows distribution and density and combines aspects of boxplots,KDE
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("Tips")

sns.violinplot(data=tips, x="day", y="total_bill", hue="sex",
               split=True, palette="Spectral")
plt.title("Total Bill Distribution by Day and Gender")
plt.show()
```

```python
#HEATMAP
#shows correlation
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

iris_data = load_iris()
iris = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)

corr = iris.corr()

sns.heatmap(corr, annot=True, cmap="PiYG")
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
#PAIRPLOT
# relation between multiple variables

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

iris_data = load_iris() 

iris["species"] = iris_data.target

sns.pairplot(iris, hue="species", palette="dark")
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

```python
#JOINTPLOT

sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter", hue="sex")
plt.suptitle("Joint Plot of Total Bill vs Tip", y=1.02)
plt.show()
```

```python
#SWARMPLOT

sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex",
              palette="YlGnBu", dodge=True)
plt.title("SWARM PLOT OF TOTAL BILL BY DAY AND GENDER")
plt.show()
```
