```python
#MATPLOTLIB
import matplotlib.pyplot as plt
```

```python
#lilne plots are ideal for visualizing trends over a continuous variable such as time

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x,y, color="green", marker="X", linestyle="--")
plt.title("Line Plot Example")
plt.xlabel("Time")
plt.ylabel("Distance covered")
plt.show()
```

```python
import matplotlib.pyplot as plt
categories = ["A", "B", "C", "D", "E", "F", "G", "H"]
values = [5, 7, 3, 8, 9, 10, 11, 11]

plt.bar(categories, values, color="teal")
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
import numpy as np
#scattterplot shows realtion between 2 varaibles often correlation
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color="orange")
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#histogram display distribution of dataset by grouping data into bins

data = np.random.randn(1000)

plt.hist(data, bins=20, color="grey", edgecolor="blue")
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#pie chart shows proportion of categories within a whole

labels = ["Nitrogen", "Oxygen", "Other Gases"]
sizes = [78, 21, 1]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=0)
plt.title("Composition of Air")
plt.show()
```

```python
#SUBPLOTS
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.plot(x, y1, color="lime")
ax1.set_title("Sine wave")

ax2.plot(x, y2, color="gold")
ax2.set_title("Cosine wave")

plt.suptitle("Sine and Cosine Waves")
plt.show()
```
