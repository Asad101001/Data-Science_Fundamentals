```python
# Average Time Between Registration and Last Purchase

#Convert dates to datetime
df['Registration_Date'] = pd.to_datetime(df['Registration_Date'])
df['Last_Purchase_Date'] = pd.to_datetime(df['Last_Purchase_Date'])

#Calculate time difference then calculate the avg
df['time_between'] = (df['Last_Purchase_Date'] - df['Registration_Date']).dt.days
average_time = df['time_between'].mean()

average_time
```

```python
# Check for missing values
missing_data = df.isnull().sum()
missing_percentage = (df.isnull().mean() * 100)
missing_data_summary = pd.DataFrame({'Missing Values': missing_data, 'Percentage': missing_percentage})
missing_data_summary = missing_data_summary.sort_values(by='Percentage', ascending=False)

# Print the summary of missing data
print(missing_data_summary)

# Visualize missing data with a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Heatmap of Missing Data')
plt.show()

```

```python
# Outlier Detetion for Annual Income

Q1 = df['Annual_Income'].quantile(0.25)
Q3 = df['Annual_Income'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Annual_Income'] < lower_bound) | (df['Annual_Income'] > upper_bound)]

outliers
```

```python
# 3. Outlier Analysis for Spending Score (using IQR)
Q1_spending = df['Spending_Score'].quantile(0.25)
Q3_spending = df['Spending_Score'].quantile(0.75)
IQR_spending = Q3_spending - Q1_spending

lower_bound_spending = Q1_spending - 1.5 * IQR_spending
upper_bound_spending = Q3_spending + 1.5 * IQR_spending
outliers_spending = df[(df['Spending_Score'] < lower_bound_spending) | (df['Spending_Score'] > upper_bound_spending)]

outliers_spending
```

```python
# Loyalty Points by Age Group

bins = [18, 25, 35, 45, 55, 65, 100]
labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '65+']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
loyalty_points_by_age_group = df.groupby('Age_Group')['Loyalty_Points'].mean().reset_index()
print(loyalty_points_by_age_group)

# Visualize Loyalty Points by Age Group (Bar plot)
sns.barplot(x='Age_Group', y='Loyalty_Points', data=loyalty_points_by_age_group)
plt.title('Loyalty Points by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Loyalty Points')
plt.show()
```

```python
# Monthly purchase trends

df['Year'] = df['Registration_Date'].dt.year
df['Month'] = df['Registration_Date'].dt.month

monthly_trends = df.groupby(['Year', 'Month'])['Purchase_Amount'].sum().reset_index()

# Combine Year and Month into a single column for easy plotting
monthly_trends['Year-Month'] = monthly_trends['Year'].astype(str) + '-' + monthly_trends['Month'].astype(str).str.zfill(2)

# Step 4: Visualize the Monthly Purchase Trends
plt.figure(figsize=(16, 8))
sns.lineplot(data=monthly_trends, x='Year-Month', y='Purchase_Amount', marker='o', color='red')
plt.title('Monthly Purchase Trends')
plt.xlabel('Month')
plt.ylabel('Total Purchase Amount')
plt.show()
```

```python
# Product popularity by age group

bins = [18, 25, 35, 45, 55, 65, 100]  # Define age range boundaries
labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '65+']  # Labels for each group
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

product_popularity = df.groupby(['Age_Group', 'Product_Purchased'])['Product_Purchased'].count().reset_index(name='Purchase_Count')

# Visualize Product Popularity by Age Group
sns.barplot(data=product_popularity, x='Age_Group', y='Purchase_Count', hue='Product_Purchased', dodge=True)
plt.title('Product Popularity by Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Number of Purchases', fontsize=12)
plt.legend(title='Product')
plt.show()
```

```python
# Monthly Purchases Analysis

df['Registration_Date'] = pd.to_datetime(df['Registration_Date'], errors='coerce')
df['Month'] = df['Registration_Date'].dt.month_name()


monthly_trends = df.groupby(['Month'])['Purchase_Amount'].sum().reset_index()

# Visualize Monthly Purchases
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_trends, x='Month', y='Purchase_Amount', marker='o', color='b')
plt.title('Monthly Purchases Analysis')
plt.xlabel('Month')
plt.ylabel('Total Purchase Amount')
plt.show()
```

```python
# Step 1: Convert 'Purchase_Date' to datetime format
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'], errors='coerce')

# Step 2: Extract the Year and Month from the Purchase Date
df['Year'] = df['Purchase_Date'].dt.year
df['Month'] = df['Purchase_Date'].dt.month_name()  # Month names like January, February, etc.

# Step 3: Define the correct order for the months
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 
               'September', 'October', 'November', 'December']

# Convert 'Month' to a categorical type with the defined order
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)

# Step 4: Group by Year and Month and calculate total purchases or total spending
# Assuming 'Purchase_Amount' is the column for the amount spent on each purchase
monthly_trends = df.groupby(['Year', 'Month'])['Purchase_Amount'].sum().reset_index()

# Combine Year and Month for easy plotting
monthly_trends['Year-Month'] = monthly_trends['Year'].astype(str) + '-' + monthly_trends['Month']

# Step 5: Visualize Monthly Purchases with Correct Month Order
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_trends, x='Year-Month', y='Purchase_Amount', marker='o', color='b')

# Customize the plot
plt.title('Monthly Purchases Analysis', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Purchase Amount', fontsize=12)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
