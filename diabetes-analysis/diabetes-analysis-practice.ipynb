```python
import pandas as pd
df = pd.read_csv("raw_diabetes_data.csv")

print ("FIRST 10 ROWS OF DATA:\n")
df.head(10)
```

```python
#Check number of rows and columns
print ("Numbersof rows:", df.shape[0])
print ("Number of columns:", df.shape[1])

#Display column names
print ("Column Names:", df.columns)
```

```python
#Check for missing values

missing_patientid = df["Patient_ID"].isnull().sum()
missing_age = df["Age"].isnull().sum()
missing_gender = df["Gender"].isnull().sum()
missing_bmi = df["BMI"].isnull().sum()
missing_glucose = df["Glucose_Level"].isnull().sum()
missing_bp = df["Blood_Pressure"].isnull().sum()
missing_insulin = df["Insulin"].isnull().sum()
missing_pregnancies = df["Pregnancies"].isnull().sum()
missing_outcome = df["Outcome"].isnull().sum()


print ("Missing values in Patient ID:", missing_patientid)
print ("Missing values in Age:", missing_age)
print ("Missing values in Gender:", missing_gender)
print ("Missing values in BMI:", missing_bmi)
print ("Missing values in Glucose_Level:", missing_glucose)
print ("Missing values in Blood_Pressure:", missing_bp)
print ("Missing values in Insulin:", missing_insulin)
print ("Missing values in Pregnancies:", missing_pregnancies)
print ("Missing values in Outcome :", missing_outcome)
```

```python
#total rows before removing duplicates
rows_beforedupes = df.shape[0]
dupess = df.duplicated().sum()
#remove duplicates
df = df.drop_duplicates()

#total rows after removing duplicates
rows_afterdupes = df.shape[0]

print ("Rows before removing duplicates:", rows_beforedupes)
print ("Rows after removing duplicates:", rows_afterdupes)
print ("Number of Duplicate Rows:", dupess)
print (df.shape)
```

```python
#Handle missing values based on logical reasoning

#BMI: median bmi grouped by gender
#Glucose_Level: mean glucose level by outcome
#Blood_Pressure: median blood pressure by Age ranges
#Insulin: median insulin by outcome
```

```python
#Fill missing BMI values by Gender Value

median_bmi_male = df[df["Gender"].str.strip() == "Male"]["BMI"].median()
median_bmi_female = df[df["Gender"].str.strip() == "Female"]["BMI"].median()

df.loc[df["Gender"].str.strip() == "Male", "BMI"] = df[df[
     "Gender"].str.strip() == "Male"]["BMI"].fillna(median_bmi_male)


df.loc[df["Gender"].str.strip() == "Female", "BMI"] = df[df[
     "Gender"].str.strip() == "Female"]["BMI"].fillna(median_bmi_female)

df.head(25)
```

```python
#Fill missing Glucose_Level values by Outcome mean

mean_glucose_no = df[df["Outcome"]== 0]["Glucose_Level"].mean()
mean_glucose_yes = df[df["Outcome"]== 1]["Glucose_Level"].mean()

df.loc[df["Outcome"]== 0, "Glucose_Level"] = df[df[
     "Outcome"] == 0]["BMI"].fillna(mean_glucose_no)

df.loc[df["Outcome"]== 1, "Glucose_Level"] = df[df[
     "Outcome"] == 1]["BMI"].fillna(mean_glucose_yes)

df.head(20)
```

```python
df["Age_Range"] = pd.cut(df["Age"], bins=[0, 30, 50, 80], labels=[
    "Young", "Middle_Age", "Old"])

median_bp_young = df[df["Age_Range"] == "Young"]["Blood_Pressure"].median
median_bp_middle = df[df["Age_Range"] == "Middle_Age"]["Blood_Pressure"].median
median_bp_old = df[df["Age_Range"] == "Old"]["Blood_Pressure"].median

df.loc[df["Age_Range"] == "Young", "Blood_Pressure"] = df[df["Age_Range"] == "Young"]["Blood_Pressure"].fillna(median_bp_young)

df.loc[df["Age_Range"] == "Middle_Age", "Blood_Pressure"] = df[df[
     "Age_Range"] == "Middle_Age"]["Blood_Pressure"].fillna(median_bp_middle)

df.loc[df["Age_Range"] == "Old", "Blood_Pressure"] = df[df[
     "Age_Range"] == "Old"]["Blood_Pressure"].fillna(median_bp_old)

df.head(15)

```

```python
#Fill missing values Insulin values using Outcome median

median_insulin_no = df[df["Outcome"] == 0]["Insulin"].median()
median_insulin_yes = df[df["Outcome"] == 1]["Insulin"].median()

df.loc[df["Outcome"]== 0, "Insulin"] = df[df[
     "Outcome"] == 0]["Insulin"].fillna(median_insulin_no)

df.loc[df["Outcome"]== 1, "Insulin"] = df[df[
     "Outcome"] == 1]["Insulin"].fillna(median_insulin_yes)

df.head(15)
```

```python
#Strip leading trailing spaces from gender

df["Gender"] = df["Gender"].str.strip()
```
