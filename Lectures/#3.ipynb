```python
import pandas as pd
import numpy as np
import random



# Set random seed for reproducibility
np.random.seed(42)

# Generate random data
n_observations = 50000

data = {
    "Order ID": [f"ORD-{i:05d}" for i in range(1, n_observations + 1)],
    "Product Category": np.random.choice(
        ["Furniture", "Office Supplies", "Technology"], n_observations
    ),
    "Sub-Category": np.random.choice(
        ["Chairs", "Tables", "Binders", "Paper", "Phones", "Accessories", "Storage", "Appliances"],
        n_observations
    ),
    "Region": np.random.choice(["East", "West", "Central", "South"], n_observations),
    "State": np.random.choice(
        ["California", "New York", "Texas", "Florida", "Illinois", "Pennsylvania"], n_observations
    ),
    "Sales": np.round(np.random.uniform(5, 1000, n_observations), 2),
    "Quantity": np.random.randint(1, 20, n_observations),
    "Discount": np.round(np.random.uniform(0, 0.5, n_observations), 2),
    "Profit": np.round(np.random.uniform(-200, 300, n_observations), 2),
    "Customer Segment": np.random.choice(
        ["Consumer", "Corporate", "Home Office", "Small Business"], n_observations
    )
}

# Create DataFrame
df_sales = pd.DataFrame(data)

# Display the first few rows of the dataset
df_sales.head()
```

```python
###############################

# Reimporting libraries and regenerating the heart disease dataset due to reset
import pandas as pd
import numpy as np



n_observations = 10000

data = {
    "Age": np.random.randint(29, 77, n_observations),
    "Sex": np.random.choice(["Male", "Female"], n_observations),
    "Chest Pain Type": np.random.choice(
        ["Typical Angina", "Atypical Angina", "Non-Anginal Pain", "Asymptomatic"],
        n_observations
    ),
    "Resting Blood Pressure (mm Hg)": np.random.randint(90, 200, n_observations),
    "Cholesterol (mg/dl)": np.random.randint(150, 400, n_observations),
    "Fasting Blood Sugar > 120 mg/dl": np.random.choice([0, 1], n_observations),
    "Resting ECG Results": np.random.choice(
        ["Normal", "ST-T Wave Abnormality", "Left Ventricular Hypertrophy"],
        n_observations
    ),
    "Max Heart Rate Achieved": np.random.randint(70, 210, n_observations),
    "Exercise-Induced Angina": np.random.choice([0, 1], n_observations),
    "Oldpeak (ST Depression Induced)": np.round(np.random.uniform(0, 6.2, n_observations), 2),
    "Slope of ST Segment": np.random.choice(["Upsloping", "Flat", "Downsloping"], n_observations),
    "Number of Major Vessels": np.random.randint(0, 4, n_observations),
    "Thalassemia": np.random.choice(["Normal", "Fixed Defect", "Reversible Defect"], n_observations),
    "Heart Disease (Target)": np.random.choice([0, 1], n_observations)  # 0: No, 1: Yes
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the dataset
df.head()
```
